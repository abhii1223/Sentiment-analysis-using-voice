```python
def get_features(filename):
    y, sr = librosa.load(filename, duration=3, offset=0.5)
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
    return mfcc
```

```python
# Load the model
model = keras.models.load_model('/content/drive/MyDrive/x/my_model.h5')
```

```python
path = ('/content/drive/MyDrive/archive (1).zip (Unzipped Files)/TESS Toronto emotional speech set data/Angry/AAF_life_angry.wav')
Audio(path)
```

```python
feature = get_features(path)
feature= np.expand_dims(feature, axis=0)
feature= np.expand_dims(feature, axis=-1)
```

```python
feature.shape
```

```python
pred=model.predict(feature)
pred=pred.argmax(axis=1)
pred = pred.astype(int).flatten()
```
